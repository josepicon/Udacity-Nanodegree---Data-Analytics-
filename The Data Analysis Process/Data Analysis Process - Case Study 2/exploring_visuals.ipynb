# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
# load datasets
import pandas as pd 
% matplotlib inline 
df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
df_08.info(),
df_18.info()
```

```python
#creating histogram for df_08 to find distribution 
df_08['greenhouse_gas_score'].plot(kind = 'hist')
```

```python
#creating histogram for df_18 to find distribution 
df_18['greenhouse_gas_score'].plot(kind = 'hist');
```

```python
#creating scatterplot for displacement (displ) and combined mpg to find correlation 

df_18.plot(x = 'displ', y= 'cmb_mpg', kind = 'scatter');
```

```python
#creating scatterplot for greenhouse gas scores and combined mpg to find correlation 

df_18.plot(x = 'greenhouse_gas_score', y= 'cmb_mpg', kind = 'scatter');
```
